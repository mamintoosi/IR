```python
sent_tok = hazm.sent_tokenize(matn)
len(sent_tok)
print(sent_tok[0:2])
```

```python
word_tok = hazm.word_tokenize(matn)
print(len(word_tok))
print(word_tok[1:20])
print(type(word_tok))
```

```python
kalemat = set()
for w in set(word_tok):
  if w.endswith('شن'):
    kalemat.add(w)
#     print(w)
print(kalemat)
print(type(kalemat))
```

```python
# loghat = {w for w in set(word_tok) if w.endswith('شن')}
loghat = [w for w in set(word_tok) if w.endswith('شن')]
print(loghat)
print(type(loghat))
```

```python
tricky = sorted([w for w in set(word_tok) if 'گل' in w or 'چمن' in w])
print(tricky)
```
